```python
import pandas as pd
import numpy as np
```

```python
class StrategyReport():

    def __init__(self,data:pd.DataFrame()):

        self.data=data
        close=self.data['c'].copy()
        self.data['return']=np.log(close/close.shift())


    def simple_return(self):

        returns=self.data['return'].copy()
        return returns.cumsum().apply(np.exp)[-1]
        

        
        

        
        

        

        
```
